```python
from urllib.request import urlretrieve
from os.path import isfile, isdir
from tqdm import tqdm
import zipfile
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import urllib.request

# Define the dataset folder and the file path
dataset_folder_path = 'dataset/'
file_path = os.path.join(dataset_folder_path, 'italy_total.csv')

# Create the dataset folder if it doesn't exist
if not isdir(dataset_folder_path):
    os.makedirs(dataset_folder_path)

# URL to download the CSV file
url = 'https://github.com/RamiKrispin/covid19Italy/raw/master/csv/italy_total.csv'

# Download the file
urllib.request.urlretrieve(url, file_path)

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the column sum of nan
print(df.isnull().sum())

# Display the dimension
print(df.shape)

# Display the first few rows of the DataFrame
print(df.head())

# Display DataFrame info
print(df.info())

# Display statistical description of the DataFrame
print(df.describe())
```
